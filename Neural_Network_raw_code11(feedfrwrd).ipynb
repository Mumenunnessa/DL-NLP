```
import numpy as np

def softmax(x):
  return 1 / (1 + np.exp(-x))

def deriv_softmax(x):
  fx = softmax(x)
  return fx * (1 - fx)

def mse_loss(y_true, y_pred):
  return ((y_true - y_pred) ** 2).mean()

class NeuralNetwork:
  def __init__(self):

    # Weights
    self.w1 = np.random.normal()
    self.w2 = np.random.normal()
    self.w3 = np.random.normal()
    self.w4 = np.random.normal()
    self.w5 = np.random.normal()
    self.w6 = np.random.normal()
    self.w7 = np.random.normal()
    self.w8 = np.random.normal()
    self.w9 = np.random.normal()
    self.w10 = np.random.normal()
    self.w11 = np.random.normal()
    self.w12 = np.random.normal()

    # Biases
    self.b1 = np.random.normal()
    self.b2 = np.random.normal()
    self.b3 = np.random.normal()
    self.b4 = np.random.normal()
    self.b5 = np.random.normal()

  def feedforward(self, x):

    h1 = softmax(self.w1 * x[0] + self.w2 * x[0] + self.b1)
    h2 = softmax(self.w3 * x[0] + self.w4 * x[1] + self.b2)
    h3 = softmax(self.w5 * x[1] + self.w6 * x[0] + self.b3)
    h4 = softmax(self.w7 * x[1] + self.w8 * x[1] + self.b4)
    o1 = softmax(self.w9 * h1 + self.w10 * h2 + self.w11 * h3 + self.w12 * h4 + self.b5)
    return o1
     

  def train(self, data, all_y_trues):
 
    lr = 0.1
    epochs = 100

    for epoch in range(epochs):
      for x, y_true in zip(data, all_y_trues):
   
        sum_h1 = self.w1 * x[0] + self.w2 * x[0] + self.b1
        h1 = softmax(sum_h1)
        sum_h2 = self.w3 * x[0] + self.w4 * x[1] + self.b2
        h2 = softmax(sum_h2)
        sum_h3 = self.w5 * x[1] + self.w6 * x[0] + self.b3
        h3 = softmax(sum_h3)
        sum_h4 = self.w7 * x[1] + self.w8 * x[1] + self.b4
        h4 = softmax(sum_h4)


        sum_o1 = self.w9 * h1 + self.w10 * h2 + self.w11 * h3 + self.w12 * h4 + self.b5
        o1 = softmax(sum_o1)
        y_pred = o1

        d_L_d_ypred = 2 * (y_true - y_pred)

        # Neuron o1
        d_ypred_d_w9 = h1 * deriv_softmax(sum_o1)
        d_ypred_d_w10 = h2 * deriv_softmax(sum_o1)
        d_ypred_d_w11 = h3 * deriv_softmax(sum_o1)
        d_ypred_d_w12 = h4 * deriv_softmax(sum_o1)
        d_ypred_d_b5 = deriv_softmax(sum_o1)

        d_ypred_d_h1 = self.w9 * deriv_softmax(sum_o1)
        d_ypred_d_h2 = self.w10 * deriv_softmax(sum_o1)
        d_ypred_d_h3 = self.w11 * deriv_softmax(sum_o1)
        d_ypred_d_h4 = self.w12 * deriv_softmax(sum_o1)

        # Neuron h1
        d_h1_d_w1 = x[0] * deriv_softmax(sum_h1)
        d_h1_d_w2 = x[0] * deriv_softmax(sum_h1)
        d_h1_d_b1 = deriv_softmax(sum_h1)

        # Neuron h2
        d_h2_d_w3 = x[0] * deriv_softmax(sum_h2)
        d_h2_d_w4 = x[1] * deriv_softmax(sum_h2)
        d_h2_d_b2 = deriv_softmax(sum_h2)
        
        # Neuron h3
        d_h3_d_w5 = x[1] * deriv_softmax(sum_h3)
        d_h3_d_w6 = x[0] * deriv_softmax(sum_h3)
        d_h3_d_b3 = deriv_softmax(sum_h3)

        # Neuron h4
        d_h4_d_w7 = x[1] * deriv_softmax(sum_h4)
        d_h4_d_w8 = x[1] * deriv_softmax(sum_h4)
        d_h4_d_b4 = deriv_softmax(sum_h4)

        # Update weights and biases

        # Neuron h1
        self.w1 = self.w1 + lr * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_w1
        self.w2 = self.w2 + lr * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_w2
        self.b1 = self.b1 + lr * d_L_d_ypred * d_ypred_d_h1 * d_h1_d_b1

        # Neuron h2
        self.w3 =  self.w3 +  lr * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_w3
        self.w4 =  self.w4 + lr * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_w4
        self.b2 =  self.b2 + lr * d_L_d_ypred * d_ypred_d_h2 * d_h2_d_b2

        # Neuron h3
        self.w5 =  self.w5 +  lr * d_L_d_ypred * d_ypred_d_h3 * d_h3_d_w5
        self.w6 =  self.w6 + lr * d_L_d_ypred * d_ypred_d_h3 * d_h3_d_w6
        self.b3 =  self.b3 + lr * d_L_d_ypred * d_ypred_d_h3 * d_h3_d_b3

        # Neuron h4
        self.w7 =  self.w7 +  lr * d_L_d_ypred * d_ypred_d_h4 * d_h4_d_w7
        self.w8 =  self.w8 + lr * d_L_d_ypred * d_ypred_d_h4 * d_h4_d_w8
        self.b4 =  self.b4 + lr * d_L_d_ypred * d_ypred_d_h4 * d_h4_d_b4

        # Neuron o1
        self.w9 =  self.w9 + lr * d_L_d_ypred * d_ypred_d_w9
        self.w10 =  self.w10 + lr * d_L_d_ypred * d_ypred_d_w10
        self.w11 =  self.w11 + lr * d_L_d_ypred * d_ypred_d_w11
        self.w12 =  self.w12 + lr * d_L_d_ypred * d_ypred_d_w12
        self.b5 =  self.b5 + lr * d_L_d_ypred * d_ypred_d_b5

      if epoch % 100 == 0:
        y_preds = np.apply_along_axis(self.feedforward, 1, data)
        loss = mse_loss(all_y_trues, y_preds)
        print("Epoch %d loss: %.3f" % (epoch, loss))

data = np.array([[0, 0],  [0, 1],   [1, 0],   [1, 1]])
all_y_trues = np.array([0, 1, 1, 0])

network = NeuralNetwork()
network.train(data, all_y_trues)

test1 = np.array([0, 0]) 
test2 = np.array([0, 1])
test3 = np.array([1, 0]) 
test4 = np.array([1, 1])  
print("test1: %.3f" % network.feedforward(test1)) 
print("test2: %.3f" % network.feedforward(test2)) 
print("test3: %.3f" % network.feedforward(test3)) 
print("test4: %.3f" % network.feedforward(test4))
```
